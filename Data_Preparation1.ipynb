## Removendo colunas e transformações de dados

```python
# Removendo colunas que não serão utilizadas:
df = df.drop(columns=['Unnamed: 0', 'name', 'category', 'goal', 'pledged', 'usd pledged'])
df.head()
```

```python
df.dtypes
```

```python
# Criando uma função para remover USD dos valores:
df['usd_goal_real'] = df['usd_goal_real'].apply(lambda x: x.strip('USD '))
df.head(10)
```

```python
# Função para remover todas strings de hora
df['launched'] = df['launched'].str.replace(' \d\d:\d\d', '', regex=True)
df.head(10)
```

```python
df.dtypes
```

```python
# Mudando a coluna usd_goal_real que está como "object" para "int64"
df['usd_goal_real'] = df['usd_goal_real'].astype('int64')
df.dtypes
```

```python
# Mudando os formatos de data de objetc para datetime
df['deadline'] =pd.to_datetime(df['deadline'], format='%d/%m/%y')
df['launched'] =pd.to_datetime(df['launched'], format='%d/%m/%y')
df.dtypes
```

```python
# Calculando o tempo que demorou para finalizar a negociação
(df['deadline']-df['launched']).dt.days
```

```python
# Adicionando uma nova coluna do data frame com esses cálculos
df['time_range'] = (df['deadline']-df['launched']).dt.days
df.head()
```

```python
# Verificando como está os dados dessa nova coluna inserida
df.dtypes
# Já consta como númerica
```

## Utilizando a técnica One hot encode, somas e médias de algumas colunas e fazendo a junção de todas as tabelas

```python
# Como temos o mesmo "ID" porém vindo investimento de outras regiões precisamos categorizar utilizando 'one hot encode'
pd.get_dummies(df_invest, columns=['backedLocation'], dtype=int)
```

```python
# Transformando esse dataset no principal agora
df_invest = pd.get_dummies(df_invest, columns=['backedLocation'], dtype=int)
```

```python
# Agora conseguimos observar os dados de uma maneira melhor, após aplicar a tecnica one hot enconde, se aparecer o mesmo ID mas em paises diferentes ficara melhor para analises
df_invest.sort_values(by = ['ID'])
```

```python
# Após esse dataframe trabalhado e ajustado para melhor visualizações dos dados, farei a soma de investimento pelo mesmo 'ID' e saber qual a media de idade que faz mais ivestimentos
df_invest.groupby(by=['ID']).agg({'age':'mean','backedLocation_BR': 'sum', 'backedLocation_GBK' : 'sum', 'backedLocation_US': 'sum'})

```

```python
# Atualizando para esse ser o novo dataframe
df_invest = df_invest.groupby(by=['ID']).agg({'age':'mean','backedLocation_BR': 'sum', 'backedLocation_GBK' : 'sum', 'backedLocation_US': 'sum'})
```

```python
df_invest.head()
```

```python
# Unindo essa base de dados ao dataframe principal
df.merge(df_invest, how =  'right', on='ID')
```

```python
# Atualizando para ser o novo dataframe
df = df.merge(df_invest, how =  'right', on='ID')
df.head()
```
